```python
import pandas as pd 
from pandas import DataFrame
import numpy as np
import random
import matplotlib.pyplot as plt
from matplotlib.pyplot import plot, show, savefig, xlim, figure, \
                ylim, legend, boxplot, setp, axes
```

```python
df2_500k = pd.read_csv('500k.csv')
```

```python
# clear current figure
plt.clf()
plt.figure(figsize=(216,288))
df2_500k.plot(x = "Subsidy", y = "OBJECTIVE")


plt.xlabel('Subsidy')
plt.ylabel('Objective Value') 
plt.title('Objective value for fixed amount of subsidy (10 year)')

plt.savefig('HP_Fixed_Subsidy_10.png', dpi=300, bbox_inches='tight')

plt.show()
```
