# 10-year flood

```python
df1 = pd.read_csv('5M_10_No_NB.csv')
df11 = pd.read_csv('5M_10_3of5_NB.csv')
df12 = pd.read_csv('5M_10_3of7_NB.csv')
df13 = pd.read_csv('5M_10_3of10_NB.csv')
```

```python
df1_count = df1.mot_year.value_counts().sort_index()
df11_count = df11.mot_year.value_counts().sort_index()
df12_count = df12.mot_year.value_counts().sort_index()
df13_count = df13.mot_year.value_counts().sort_index()
df13_count
```

```python
# clear current figure
plt.clf()
fig = figure(figsize=(15, 8))
plt.title('Number of motivated moves of the following year (10 year)')

df1_count.cumsum().plot()
df11_count.cumsum().plot()
df12_count.cumsum().plot()
df13_count.cumsum().plot()

plt.xlim([0,80])

plt.ylabel('Number of moves')
plt.xlabel('Relocation year')

plt.legend(labels=['No neighborhood effect','Motivated move + 3 out 5 neighbors have moved','Motivated move + 3 out 7 neighbors have moved','Motivated move + 3 out 10 neighbors have moved']) 

plt.show()
```

```python
# In general, year 3 has the least motivated (NB and sub = 0) moves

df1_count_ori = df1.ori_year.value_counts().sort_index()
df11_count_ori = df11.ori_year.value_counts().sort_index()
df12_count_ori = df12.ori_year.value_counts().sort_index()
df13_count_ori = df13.ori_year.value_counts().sort_index()
df1_count_ori
```

```python
fig = figure(figsize=(15, 8))
plot1 = plt.figure(1)
df1_count.cumsum().plot()
# df1_count_ori.cumsum().plot()
plt.xlim([0,5])
plt.ylabel('Number of moves')
plt.xlabel('Relocation year')
plt.legend(labels=['No neighborhood effect'])

fig = figure(figsize=(15, 8))
plot2 = plt.figure(2)
df11_count.cumsum().plot()
df11_count_ori.cumsum().plot()
plt.xlim([0,5])
plt.ylabel('Number of moves')
plt.xlabel('Relocation year')
plt.legend(labels=['Motivated move + 3 out 5 neighbors have moved','No neighborhood effect'])

fig = figure(figsize=(15, 8))
plot3 = plt.figure(3)
df12_count.cumsum().plot()
df12_count_ori.cumsum().plot()
plt.xlim([0,5])
plt.ylabel('Number of moves')
plt.xlabel('Relocation year')
plt.legend(labels=['Motivated move + 3 out 7 neighbors have moved','No neighborhood effect'])

fig = figure(figsize=(15, 8))
plot4 = plt.figure(4)
df13_count.cumsum().plot()
df13_count_ori.cumsum().plot()
plt.xlim([0,5])
plt.ylabel('Number of moves')
plt.xlabel('Relocation year')
plt.legend(labels=['Motivated move + 3 out 10 neighbors have moved','No neighborhood effect'])

plt.show()
```
